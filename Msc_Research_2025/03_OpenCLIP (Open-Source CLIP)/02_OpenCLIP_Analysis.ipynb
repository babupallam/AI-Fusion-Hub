## **3. Understanding OpenCLIP**
### **What is OpenCLIP?**
OpenCLIP is an open-source implementation of OpenAI’s CLIP model, created to expand accessibility and further research in vision-language pretraining. While OpenAI originally developed CLIP, OpenCLIP was created by researchers from ML Foundations, allowing broader access to the technology.

### **Key Features of OpenCLIP**
- Open-source alternative to CLIP with additional model architectures.
- Trained on diverse datasets like **LAION-2B** (a dataset containing 2 billion image-text pairs).
- Supports multiple vision architectures such as **ViT (Vision Transformers)** and **ConvNext**.
- Provides zero-shot classification capabilities across multiple datasets.

### **Comparison: OpenAI CLIP vs OpenCLIP**
| Feature | OpenAI CLIP | OpenCLIP |
|---------|------------|----------|
| **Developer** | OpenAI | ML Foundations |
| **Availability** | Closed-source | Open-source |
| **Pretraining Dataset** | WebImageText (WIT) | LAION-400M, LAION-2B, DataComp-1B |
| **Models** | ViT-based | ViT, ConvNext, CoCa |
| **Fine-Tuning** | Limited | Supports fine-tuning |
| **Performance** | Stronger on general datasets | Competitive with additional flexibility |

### **Significance of OpenCLIP**
OpenCLIP has enabled researchers and developers to fine-tune and use CLIP models without restrictions. It has been used for tasks such as:
- **Zero-shot image classification** (classifying images without additional training).

## Problem:  The current text prompts are manually defined, but they should be generated dynamically based on the input image rather than predefined


## Solution: Automatically Generate Text Prompts for the Image


To generate text prompts based on the image, we can use BLIP (Bootstrapped Language-Image Pretraining), a model that generates textual descriptions of an image. We will:

  1. Use OpenCLIP for classification
  2. Use BLIP to generate a textual description of the image
  3. Use the generated description as a prompt for OpenCLIP

````python
# Step 1: Install necessary libraries (if not installed)
# This installs OpenCLIP, transformers, and other required libraries.
# Run this command in the terminal or Jupyter Notebook if needed.
# !pip install open_clip_torch transformers torch torchvision pillow matplotlib

# Step 2: Import required libraries
import torch  # PyTorch for model operations and tensor calculations
import matplotlib.pyplot as plt  # Visualization library for displaying images and graphs
from PIL import Image  # Handling image processing
import open_clip  # OpenCLIP model for vision-language understanding
from transformers import BlipProcessor, BlipForConditionalGeneration  # BLIP model for generating captions

````

````python
# Step 3: Load a pretrained OpenCLIP model
# OpenCLIP is used for encoding both images and text into a shared embedding space.
# 'ViT-B-32' is a Vision Transformer (ViT) model trained with CLIP on LAION-2B dataset.
model, preprocess, tokenizer = open_clip.create_model_and_transforms('ViT-B-32', pretrained='laion2b_s34b_b79k')

# Step 4: Load the BLIP model for automatic caption generation
# BLIP (Bootstrapped Language-Image Pretraining) is used to generate a caption for an input image.
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

````

````python
# Step 5: Load and preprocess the image
# The image is loaded using PIL (Pillow) and converted to RGB format.
image_path = "test2.jpg"  # Change to your image path
image = Image.open(image_path).convert("RGB")

# Step 6: Generate a text prompt dynamically using BLIP
# The BLIP model processes the image and generates a caption.
inputs = blip_processor(image, return_tensors="pt")  # Convert image to tensor format
with torch.no_grad():  # Disables gradient calculations for faster inference
    caption_ids = blip_model.generate(**inputs)  # Generate caption tokens
generated_prompt = blip_processor.batch_decode(caption_ids, skip_special_tokens=True)[0]  # Convert tokens to text

# Display the generated text prompt
print(f"Generated Text Prompt: {generated_prompt}")

````

````python
# Step 7: Tokenize the generated text prompt
# The text prompt is converted into tokenized format for OpenCLIP processing.
text = open_clip.tokenize([generated_prompt])  # Tokenize the generated caption

# Step 8: Preprocess the image for OpenCLIP
# The image is transformed into the format required by the OpenCLIP model.
image_tensor = preprocess(image).unsqueeze(0)  # Add batch dimension for model input

# Step 9: Encode image and text features
# Compute image and text embeddings using OpenCLIP.
# Using `torch.cuda.amp.autocast()` ensures mixed-precision calculations for efficiency.
with torch.no_grad(), torch.cuda.amp.autocast():
    image_features = model.encode_image(image_tensor)  # Encode image into feature vector
    text_features = model.encode_text(text)  # Encode text into feature vector

````

````python
# Step 10: Normalize the embeddings
# Normalization ensures embeddings have unit norm (magnitude = 1) for accurate similarity comparison.
image_features /= image_features.norm(dim=-1, keepdim=True)
text_features /= text_features.norm(dim=-1, keepdim=True)

# Step 11: Compute similarity scores
# The similarity score is computed as the dot product of the normalized image and text embeddings.
# Higher similarity means the generated text closely matches the image content.
text_probs = (100.0 * image_features @ text_features.T).softmax(dim=-1)

# Step 12: Print the similarity score
# The probability distribution over text labels is printed.
print("Label probabilities:", text_probs)  # Displays similarity score between image and generated caption

````

````python
# Step 13: Visualization
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # Create two subplots for image and similarity score

# Display the image with the generated caption
ax[0].imshow(image)  # Show the image
ax[0].axis("off")  # Hide axes for better visualization
ax[0].set_title(f"Generated Caption:\n{generated_prompt}", fontsize=12)  # Add caption text as title

# Display the similarity score as a bar graph
ax[1].bar(["Generated Prompt"], text_probs[0].cpu().numpy(), color="blue")  # Bar graph for similarity
ax[1].set_ylim(0, 1)  # Limit y-axis to 0-1 (probability range)
ax[1].set_ylabel("Similarity Score")  # Label for y-axis
ax[1].set_title("OpenCLIP Prediction Confidence")  # Title for the bar graph

# Show the visualization
plt.show()
````

````python
# Step 1: Install necessary libraries (if not installed)
# pip install open_clip_torch transformers torch torchvision pillow matplotlib numpy seaborn

# Step 2: Import required libraries
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import open_clip
from transformers import BlipProcessor, BlipForConditionalGeneration
import torchvision.transforms as T

````

````python
# Step 3: Load OpenCLIP Model
print("\n🔹 Loading OpenCLIP Model...\n")
model, preprocess, tokenizer = open_clip.create_model_and_transforms('ViT-B-32', pretrained='laion2b_s34b_b79k')
print("✅ OpenCLIP Model Loaded Successfully!\n")

# Step 4: Load BLIP models for caption generation
print("\n🔹 Loading BLIP Models (Base & Large)...\n")
blip_models = {
    "base": BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base"),
    "large": BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
}
blip_processors = {
    "base": BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base"),
    "large": BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
}
print("✅ BLIP Models Loaded Successfully!\n")

````

````python
# Step 5: Load and preprocess the image
image_path = "test2.jpg"  # Change this if needed
image = Image.open(image_path).convert("RGB")

# Step 6: Generate captions using different BLIP models
captions = {}
print("\n🔹 Generating Captions using BLIP Models...\n")
for model_name, blip_model in blip_models.items():
    processor = blip_processors[model_name]
    inputs = processor(image, return_tensors="pt")
    with torch.no_grad():
        caption_ids = blip_model.generate(**inputs)
    captions[model_name] = processor.batch_decode(caption_ids, skip_special_tokens=True)[0]
    print(f"📌 {model_name.upper()} Caption: {captions[model_name]}")
print("✅ Captions Generated Successfully!\n")

````

````python
# Step 7: Tokenize the generated captions for OpenCLIP
tokenized_captions = {name: open_clip.tokenize([text]) for name, text in captions.items()}

# Step 8: Preprocess the image for OpenCLIP
image_tensor = preprocess(image).unsqueeze(0)

# Step 9: Encode image and text features using OpenCLIP
print("\n🔹 Computing Image & Text Features using OpenCLIP...\n")
image_features = model.encode_image(image_tensor)
text_features = {name: model.encode_text(text) for name, text in tokenized_captions.items()}
print("✅ Features Computed Successfully!\n")

````

````python
# Step 10: Normalize the embeddings
image_features /= image_features.norm(dim=-1, keepdim=True)
for name in text_features:
    text_features[name] /= text_features[name].norm(dim=-1, keepdim=True)

# Step 11: Compute similarity scores for each generated caption
text_probs = {name: (100.0 * image_features @ text_features[name].T).softmax(dim=-1) for name in text_features}

# Step 12: Print similarity scores
print("\n🔹 OpenCLIP Similarity Scores:\n")
for model_name, score in text_probs.items():
    print(f"✅ {model_name.upper()} Similarity Score: {score.item():.4f}")
print("✅ Similarity Scores Computed Successfully!\n")

````

````python
# Step 13: Visualization of Captions & Similarity Scores
print("\n🔹 Generating Visualization...\n")
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Display the original image with Base Caption
ax[0].imshow(image)
ax[0].axis("off")
ax[0].set_title(f"📜 Caption (Base):\n{captions['base']}", fontsize=12)

# Display the original image with Large Caption
ax[1].imshow(image)
ax[1].axis("off")
ax[1].set_title(f"📜 Caption (Large):\n{captions['large']}", fontsize=12)

# Display similarity scores as a bar chart
bar_labels = list(text_probs.keys())
bar_values = [score.item() for score in text_probs.values()]
ax[2].bar(bar_labels, bar_values, color=["blue", "green"])
ax[2].set_ylim(0, 1)
ax[2].set_ylabel("🔍 Similarity Score")
ax[2].set_title("OpenCLIP Confidence")

plt.show()
print("✅ Visualization Generated Successfully!\n")

````

````python
# Step 14: Generate Attention Heatmaps
def generate_attention_heatmap(image, model_type="base"):
    """Generates an attention heatmap overlay for the given image."""
    processor = blip_processors[model_type]
    model = blip_models[model_type]

    # Process the image using BLIP's vision encoder
    inputs = processor(image, return_tensors="pt")

    with torch.no_grad():
        encoder_outputs = model.vision_model(**inputs)  # Extract vision features

    attentions = encoder_outputs.last_hidden_state  # Extract final hidden states

    if attentions is None or attentions.shape[1] == 1:
        print(f"❌ No attention data available for {model_type}. Skipping heatmap.\n")
        return

    # Compute mean attention over the last layers
    attn_map = attentions.mean(dim=1).cpu().numpy()
    attn_map = np.array(Image.fromarray(attn_map[-1]).resize(image.size, resample=Image.BILINEAR))

    # Overlay heatmap on image
    transform = T.Compose([T.Resize((image.size[1], image.size[0])), T.ToTensor()])
    img_tensor = transform(image)

    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    ax.imshow(img_tensor.permute(1, 2, 0))
    ax.imshow(attn_map, cmap="jet", alpha=0.5)
    ax.axis("off")
    ax.set_title(f"🔥 Attention Heatmap ({model_type})")
    plt.show()

# Generate heatmaps for both BLIP models
print("\n🔹 Generating Attention Heatmaps...\n")
generate_attention_heatmap(image, model_type="base")
generate_attention_heatmap(image, model_type="large")
print("✅ Heatmaps Generated Successfully!\n")
````
